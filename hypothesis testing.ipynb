```python
def remove_outliers( columns_list,df ):
    
    "Note- use this func only after imputation, as nan value will be filter also if apply before imputation."
    
    new_df =pd.DataFrame()
    for column_name in columns_list:
        
        Q1 = df[column_name].quantile(0.25)
        Q3 = df[column_name].quantile(0.75)
        IQR = Q3 - Q1
        filter = (df[column_name] >= Q1 - 1.5 * IQR) & (df[column_name] <= Q3 + 1.5 *IQR)
        new_df=df.loc[filter]
        
    return new_df
```

```python
import scipy.stats as stats
observed_matrix = pd.crosstab(df["Goal Type: Squats"], df.repeat_fundraiser_label, margins = True)

observed=observed_matrix.iloc[:-1,:-1]

observed
```

```python
chi2, p_value, degree_of_freedom, expected  =stats.chi2_contingency(observed= observed)

rounded_p_value = round(p_value,6)

print("chi2:",chi2)
print("p_value:",rounded_p_value)
print("degree_of_freedom:",degree_of_freedom)
print("expected:\n",expected)
```

```python
critical_value = 0.05
```

```python
for column_name in columns_for_testing:
    observed_matrix = pd.crosstab(df[column_name], df.repeat_fundraiser_label, margins = True)

    observed=observed_matrix.iloc[:-1,:-1]
    
    chi2, p_value, degree_of_freedom, expected  =stats.chi2_contingency(observed= observed)

    rounded_p_value = round(p_value,6)

    if rounded_p_value<critical_value:    
        print("columns_name:",column_name)
        print("chi2:",chi2)
        print("p_value:",rounded_p_value)
        print("degree_of_freedom:",degree_of_freedom)
        print("expected:\n",expected)
        print("....................................................")
```
